```python
#Exploring one sample file
file_number = '0001'
file_path = 'CSV/{}.csv'.format(file_number)
```

```python
df = pd.read_csv(file_path)
print(df.head())
df.shape
```

```python
def FSC_norm(dataframe):
    dataframe['FS Lin'] -= float(dataframe['FS Lin'].mean())
    dataframe['FS Lin'] = dataframe['FS Lin']/dataframe['FS Lin'].std()*.1
    return dataframe
df = FSC_norm(df)
```

```python
df.head()
```

```python
def construct_row(file_df):
    return df.mean()
```

```python
#Data Preprocessing for all sample files 

#Averaging all features in each file to create 7 cells per file


num_patients = 179
file_path = 'CSV/{}.csv'
rows = []

for i in range(num_patients):
    
    print(i)
    file_nums = [j for j in range(i*8+1,i*8+9)]
    patient_data = []
    count = 1
    
    for j in file_nums:
        
        file_name = str(j)
        
        while len(file_name)< 4:
            file_name = '0' + file_name
        print("Reading file: {}".format(file_path.format(file_name)))
        
        #read in the sample file
        df = pd.read_csv(file_path.format(file_name))
        
        #normalize FSC measurements
        df = FSC_norm(df)
        
        #get the rows
        row = construct_row(df).rename({'FS Lin': 'FS Lin {}'.format(count), 'SS Log': 'SS Log {}'.format(count),'FL1 Log': 'FL1 Log {}'.format(count), 'FL2 Log': 'FL2 Log {}'.format(count),
                                        'FL3 Log': 'FL3 Log {}'.format(count), 'FL4 Log': 'FL4 Log {}'.format(count), 'FL5 Log': 'FL5 Log {}'.format(count)})
     
       
        patient_data.append(row)
        
        count += 1
    
    row = pd.concat(patient_data, axis=0)
    #rows.append(row)
    row_df = pd.DataFrame(data=row.to_frame().T)
    rows.append(row_df)
```

```python
patient_data = pd.concat(rows, axis=0).reset_index()
patient_data.head()
```

```python
patient_data = patient_data.drop(['index'], axis=1)
patient_data.head()
```

```python
patient_data.shape
```

```python
#training data
save_path = 'data/train_mean_features.csv'
patient_data.to_csv(save_path,header=True)
```

```python
#Getting test data 

start = 179
end = 359
file_path = 'CSV/{}.csv'
rows = []

for i in range(start, end):
    
    print(i)
    file_nums = [j for j in range(i*8+1,i*8+9)]
    patient_data = []
    count = 1
    
    for j in file_nums:
        
        file_name = str(j)
        while len(file_name)< 4:
            file_name = '0' + file_name
        print("Reading file: {}".format(file_path.format(file_name)))
        
        #read in the sample file
        df = pd.read_csv(file_path.format(file_name))
        

        #normalize FSC
        df = FSC_norm(df)
        
        #get the rows
        row = construct_row(df).rename({'FS Lin': 'FS Lin {}'.format(count), 'SS Log': 'SS Log {}'.format(count),'FL1 Log': 'FL1 Log {}'.format(count), 'FL2 Log': 'FL2 Log {}'.format(count),
                                        'FL3 Log': 'FL3 Log {}'.format(count), 'FL4 Log': 'FL4 Log {}'.format(count), 'FL5 Log': 'FL5 Log {}'.format(count)})
     
       
        patient_data.append(row)
        
        count += 1
    
    row = pd.concat(patient_data, axis=0)
    #rows.append(row)
    row_df = pd.DataFrame(data=row.to_frame().T)
    rows.append(row_df)
```

```python
label_map = {'aml': 1,
            'normal': 0}
```

```python
samples = 179
for i in range(samples):
    set_label = set(classification_data[i*8:i*8+8].Label)
    if len(set_label) != 1:
        print("Error in sample {}".format(i+1))
    train_data.loc[i,'label'] = label_map[list(set_label)[0]]
    
```
